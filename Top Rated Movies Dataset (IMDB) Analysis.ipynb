```python
movies_pop = df['popularity']
Q3 = movies_pop.quantile(0.75)
Q1 = movies_pop.quantile(0.25)
IQR = Q3-Q1
LW = Q1 - (1.5*IQR)
UW = Q3 + (1.5*IQR)
pop_outliers = movies_pop[(movies_pop <LW) | (movies_pop >UW)]
pop_outliers
```

```python
pop_filtered = movies_pop[(movies_pop >LW) & (movies_pop >UW)]
pop_filtered
```
